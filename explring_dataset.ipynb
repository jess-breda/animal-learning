```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import pathlib 
```

```python
DATA_PATH = pathlib.Path("X:\jbreda\learning_project\data")
F_NAME = "rat_behavior.csv"
ANIMAL_ID = "W080"
```

#### Load CSV -> DF

```python
DATA_PATH/F_NAME
```

```python
RAT_DF = pd.read_csv(DATA_PATH/F_NAME)
```

```python
RAT_DF.head()
```

```python
W080_df = RAT_DF.query('subject_id == @ANIMAL_ID').reset_index(drop=True)
```

## Sessions & Stage

```python
W080_df.head()
```

```python
ax = (W080_df.groupby("training_stage")
 .max()
 .session
 .plot.bar(
    title=f"{ANIMAL_ID} N sessions per stage",
    figsize=(5,3),
    color = 'gray'
    )
)
for p in ax.patches:
    ax.annotate(str(p.get_height()), 
                (p.get_x() * 1, 
                 p.get_height() * 1.005))

sns.despine()
plt.tight_layout()
```

```python
fig,ax = plt.subplots(figsize = (8,4))
sns.lineplot(
    data=W080_df, x="session", y ="hit", 
    hue="training_stage", errorbar=['ci', False]
    )

plt.grid()
plt.axhline(y=0.5,zorder=1, color='k')
plt.ylim(0,1.1)
ax.set(title=f"{ANIMAL_ID} Hit Rate Per Session ~ Training Stage")
```

```python
fig,ax = plt.subplots(figsize = (8,4))
sns.lineplot(W080_df["s_a"].isnull(),ax=ax,label="No Sounds")
sns.lineplot(W080_df.training_stage, ax=ax,label="Training Stage")
_ = ax.set(title=f"{ANIMAL_ID} Sounds and Training Stage",
           ylabel="Is Null Bool || Training Stage",
           xlabel= "Trials"
        )
```
