# Visuals

## Import

```python
import violmulti.utils.save_load as save_load
from violmulti.visualizations.ssm_visualizer import (
    visualize_v2,
    plot_log_posteriors_by_inits,
)


experiment_dir = save_load.get_experiment_dir(experiment_name)

lls = save_load.load_multi_model_log_likelihoods(experiment_name)

lps = save_load.load_multi_model_log_posteriors(experiment_name)

best_init = lls.argmax()

print(f"Best Inititalization: {best_init}")
```

```python
best_fit_model = save_load.load_model_from_pickle(
    animal_id="Ws",
    n_states=2,
    model_name="mega_glmhmm",
    n_fold=0,
    n_init=best_init,
    model_path=experiment_dir / "models",
)

X, y = save_load.load_data_and_labels_from_parquet(
    animal_id="Ws",
    model_name="mega_glmhmm",
    n_fold=0,
    data_path=experiment_dir / "data",
)

feature_names = [col for col in X.columns if col != "session"]
```

### Best Model

```python
visualize_v2(
    best_fit_model,
    animal_id="All Animals",
    feature_names=feature_names,
    seed=18,
)
```

### Other random model for comparison

```python
model_2 = save_load.load_model_from_pickle(
    animal_id="Ws",
    n_states=2,
    model_name="mega_glmhmm",
    n_fold=0,
    n_init=7,
    model_path=experiment_dir / "models",
)
```
